# 2017 versus 2021 by year

## SRER

```python
import numpy as np
import matplotlib.pyplot as plt
import os
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("SRER_2017_training_bi.csv")
df.isnull().sum()
df.dropna(axis=0,inplace=True)
data = pd.read_csv("SRER21_dataset_v1.csv")
data.isnull().sum()
data.dropna(axis=0,inplace=True)
print(df.shape)
print(data.shape)
```

```python
df1 = df.assign(Year=2017)
df2 = data.assign(Year=2021)
df3 = pd.concat((df1,df2), ignore_index=True)
```

```python
my_colors = ["#f39e53", "#8B5FBF"]

sns.set_palette(my_colors)
```

### Variable Graphs

```python
category = ["ARVI_mean","ARVI_med","ARVI_max","EVI_mean","EVI_med","EVI_max","NDVI_mean","NDVI_med","NDVI_max","SAVI_mean","SAVI_med","SAVI_max"]


for i in category:
    plt.figure(figsize=(8,6))
    plt.ylabel("Number of Records")
    sns.boxplot(y=i, x='Year', data=df3, showfliers=False,hue = 'Year')
```

## JORN

```python
df2 = pd.read_csv("JORN17_dataset_v2.csv")
df2.isnull().sum()
df2.dropna(axis=0,inplace=True)
data2 = pd.read_csv("JORN21_dataset_v1.csv")
data2.isnull().sum()
data2.dropna(axis=0,inplace=True)
print(df2.shape)
print(data2.shape)
```

```python
df11 = df2.assign(Year=2017)
df21 = data2.assign(Year=2021)
df31 = pd.concat((df11,df21), ignore_index=True)
```

```python
my_colors = ["#f39e53", "#8B5FBF"]

sns.set_palette(my_colors)
```

### Variable Graphs

```python
category = ["ARVI_mean","ARVI_med","ARVI_max","EVI_mean","EVI_med","EVI_max","NDVI_mean","NDVI_med","NDVI_max","SAVI_mean","SAVI_med","SAVI_max"]


for i in category:
    plt.figure(figsize=(8,6))
    plt.ylabel("Number of Records")
    sns.boxplot(y=i, x='Year', data=df31, showfliers=False,hue = 'Year')
```
