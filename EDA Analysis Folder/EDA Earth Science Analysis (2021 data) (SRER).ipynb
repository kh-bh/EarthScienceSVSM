```python
totalarea = 0
woodyarea = 0
nonwoodyarea = 0
 
w = 0
nw = 0

shapearea = data["Shape_Area"]
vegclassi = data["Veg_class"]

for x in data.index:
    totalarea += shapearea[x]
    if vegclassi[x] == "woody":
        woodyarea += shapearea[x]
        w += 1
    if vegclassi[x] == "non-woody":
        nonwoodyarea += shapearea[x]
        nw += 1
print(woodyarea)
print(nonwoodyarea)
print(totalarea)

print(w)
print(nw)

FWC = woodyarea / totalarea * 100

print(FWC)
```
