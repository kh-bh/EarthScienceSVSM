# 2017 Versus 2021 (classified by Woody and Non-woody and Year)

## SRER

```python
import numpy as np
import matplotlib.pyplot as plt
import os
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("SRER_2017_training_bi.csv")
df.isnull().sum()
df.dropna(axis=0,inplace=True)
data = pd.read_csv("SRER21_dataset_v1.csv")
data.isnull().sum()
data.dropna(axis=0,inplace=True)
print(df.shape)
print(data.shape)
```

```python
df.isnull().sum()
data.isnull().sum()
```

```python
df1 = df.assign(Year=2017)
```

```python
df2 = data.assign(Year=2021)
```

```python
df3 = pd.concat((df1,df2), ignore_index=True)
```

```python
#setting up colors

my_colors = ["#f39e53", "#8B5FBF"]


sns.set_palette(my_colors)
```

### Variable Graphs

```python
category = ["ARVI_mean","ARVI_med","ARVI_max","EVI_mean","EVI_med","EVI_max","NDVI_mean","NDVI_med","NDVI_max","SAVI_mean","SAVI_med","SAVI_max"]


for i in category:
    plt.figure(figsize=(8,6))
    plt.ylabel("Number of Records")
    sns.boxplot(y=i, x='Veg_class', hue='Year', data=df3, showfliers=False)
```
